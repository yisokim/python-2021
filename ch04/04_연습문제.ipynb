```python
# 1) 연도를 입력으로 받아 윤년인지 아닌지를 출력하는 프로그램을 작성하시오. 윤년은 연도가 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때이다.예를들어, 2012년은 4의 배수라서 윤년이지만, 1900년은 4의 배수이지만, 100의 배수이기 때문에 윤년이 아니다. 하지만, 2000년은 400의 배수이기 때문에 윤년이다.)

year = int(input('연도를 입력하시오.'))

if year % 4 == 0 and year % 100 != 0 or year % 400 == 0: print('윤년')
else: print('윤년이 아닙니다.')

##############################

if year % 400 == 0:
    print(f'{year} is leap year.')
elif year % 100 == 0:
    print(f'{year} is leap year.')
if year % 400 == 0:
    print(f'{year} is leap year.')
```

```python
# 1) 5이상 9이하의 홀수를 입력받아 다이아몬드 형태의 별을 출력하는 프로그램을 작성하시오.

# x = int(input('줄 수를 입력하세요'))
# for i in range(1, x+1):
#     for k in range(x+1-i):
#         print(' ', end='')
#     for k in range(2*i - 1):
#         print('+', end='')
#     print()

###################

import math # math 수학 함수를 쓰고 싶을 때
N = int(input('홀수 입력'))
half = math.ceil(N/2)
for i in range(1, half + 1):
    for k in range(half - i):
        print(' ', sep='', end='')
    for k in range(2*i-1):
        print('+', sep='', end='')
        print()
    for i in range(1, half):
        for k in range(i):
        print(' ', sep='', end='')
    for k in range(2*(half-i)-1):
        print('+', sep='', end='')
    print()
```

```python
# 2) 디지털 시계에 하루동안(00:00~23:59) 3이 표시되는 시간을 초로 환산하면 총 몇 초(second)일까요?- 디지털 시계는 하루동안 다음과 같이 시:분(00:00~23:59)으로 표시됨.

total_sec = 0
for hour in range(24):
    for min in range(60):
        time = str(hour) + str(min)
        if '3' in time:
            total_sec += 60
    print(total_sec)
```

```python
# 4) 자기 자신을 제외한 모든 양의 약수들의 합이 자기 자신이 되는 자연수를 완전수라고 한다. 예를 들면, 6과 28은 완전수이다. 6=1+2+3
# # 1,2,3은 각각 6의 약수 28=1+2+4+7+14   # 1,2,4,7,14는 각각 28의 약수. 입력으로 자연수 N(10,000 이하)을 받고, 출력으로 N 이하의 모든 완전수를 출력하는 코드를 작성하시오.

num = int(input('숫자 입력'))

for i in range(1, num+1):   # 1부터 num까지
    sum = 0
for j in range(1,i):        # 1부터 i까지
    if i%j == 0:
        sum += j            # 약수들의 합을 구함
    if i == sum:
        print(i)

####################

# 약수 찾기 - ex) 12
num = 12
for i in range(1,num+1):
    if num % i == 0:
        print(i)
# 자기 자신을 제외한 약수의 합
sum = 0
for i in range(1, num):
    if num % i == 0:
        sum += 1
    print(sum)

for num in range(2,10001):
    sum = 0
    for i in range(1, num):
        if num % i == 0:
            sum += i 
        if sum == num:
            print(num)
```
